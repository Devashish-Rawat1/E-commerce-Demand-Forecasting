```python
df = pd.read_csv("../data/processed/featured_data.csv", parse_dates=["date"])
```

```python

df = df.sort_values("date").reset_index(drop=True)

# split date (last ~20% as validation)
split_date = df["date"].quantile(0.80)

train_df = df[df["date"] < split_date]
val_df   = df[df["date"] >= split_date]

print("Train period:", train_df["date"].min(), "to", train_df["date"].max())
print("Val period:", val_df["date"].min(), "to", val_df["date"].max())
```
